```python
import pandas as pd
import os #列出所有檔案名稱
```

```python
filelist = os.listdir('./res_gossiping')
```

```python
data = []
for textfile in filelist:
    if textfile == '.ipynb_checkpoints':
        continue
    with open('./res_gossiping/%s'%(textfile),'r',encoding = 'utf-8') as f:
            datastr = f.read().split('---split---')[1]
            data_row = datastr.split('\n')[1:-1]
            data.append(data_row)
#             print(data_row)
#             print(datastr)
#         print('===============================')
```

```python
columns = ['推','噓','分數','作者','標題','時間']
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
def tmpfilter(s): #文字過濾器
    result = s.split(": ")[1]
    return result
```

```python
tmpfilter('推: 0')
```

```python
df_copied = df.copy() #備份dataframe
```

```python
columns
```

```python
for i in columns:
    df_copied[i] = df_copied[i].apply(tmpfilter)
df_copied
```

```python
df_copied2 = df.copy()
```

```python
for c in columns:
    df_copied2[c] = df_copied2[c].apply(lambda s:s.split(': ')[1])
df_copied2
```

```python
df_copied2.to_csv('./ptt.csv',index=False)
```
