```python
import jieba
import os


jieba.load_userdict('./mytest.txt')
```

```python
filelist = os.listdir('./res_gossiping')
```

```python
'''把抓下來的文章開啟後,再進行斷詞'''
sb = ''
for textfile in filelist:
    if textfile == '.ipynb_checkpoints': #jupyter notebook有這個隱藏檔,把他跳過
        continue
    with open('./res_gossiping/%s'%(textfile),'r',encoding = 'utf-8') as f:
        sb += f.read().split('---split---')[0] #因為要各個文章的內容,方便啟用把每個文章都裝在一起
```

```python
s_cut  = jieba.cut(sb)
```

```python
wc = {}
for w in s_cut:
    if len(w)<2: #切割出來的字如果長度小於2,則跳過
        continue
    if w not in wc: #如果w不在原有清單就為1次,如果有就加1次
        wc[w] = 1
    else:           
        wc[w] += 1
wc
```

```python
wc_list =[(w,wc[w]) for w in wc]
wc_list
```

```python
wc_list = []  #寫法同上,把他當成tuple來做
for w in wc:
    tmpdata = tuple((w,wc[w]))
    wc_list.append(tmpdata)
wc_list
```

```python
wc_list.sort(key=lambda t :t[1],reverse=True)
wc_list
```
